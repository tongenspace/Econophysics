```python
def discrete_laplacian(M):
    L = -4*M
    L += np.roll(M, (0,-1), (0,1))
    L += np.roll(M, (0,+1), (0,1))
    L += np.roll(M, (-1,0), (0,1)) 
    L += np.roll(M, (+1,0), (0,1)) 
    
    return L
def step(A, B,C, D, gamma, delta_t):

    LA = discrete_laplacian(A)
    LB = discrete_laplacian(B)
    LC = discrete_laplacian(C)
    
    diff_A = (D*LA - gamma*A*B) * delta_t
    diff_B = (D*LB - gamma*A*B) * delta_t
    diff_C = (D*LC + gamma*A*B) * delta_t
    
    A += diff_A
    B += diff_B
    C += diff_C
    
    return A, B, C
def initial(N, random_influence=0.2):
   
    
    A = np.zeros((N,N))
    B = np.zeros((N,N))
    N2 = N//2
    
    A[int(N2/4), :] = 0.50
    B[:, int(3*N2/4)] = 0.25
    C = np.zeros((N,N))
    
    return A, B, C
def draw(A,B,C):

    fig, ax = plt.subplots(1,3,figsize=(9.65,6))

    ax[0].imshow(A, cmap='Greys')
    ax[1].imshow(B, cmap='Greys')
    ax[2].imshow(C, cmap='Greys')

    ax[0].set_title('A')
    ax[1].set_title('B')
    ax[2].set_title('C')
    
    ax[0].axis('off')
    ax[1].axis('off')
    ax[2].axis('off')
```
