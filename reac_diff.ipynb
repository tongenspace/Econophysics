```python
def discrete_laplacian(M):
    L = -4*M
    L += np.roll(M, (0,-1), (0,1)) 
    L += np.roll(M, (0,+1), (0,1)) 
    L += np.roll(M, (-1,0), (0,1)) 
    L += np.roll(M, (+1,0), (0,1)) 
    
    return L
def gray_step(A, B, DA, DB, f, k, delta_t):
    
    LA = discrete_laplacian(A)
    LB = discrete_laplacian(B)
    
    diff_A = (DA*LA - A*B**2 + f*(1-A)) * delta_t
    diff_B = (DB*LB + A*B**2 - (k+f)*B) * delta_t
    
    A += diff_A
    B += diff_B
    
    return A, B

def initial(N):

    A = np.zeros((N,N))
    
    B = np.zeros((N,N))
    #
    N2 = N//2
    r = int(N/10.0)
    
    A[N2-r:N2+r, N2-r:N2+r] = 0.50
    B[N2-r:N2+r, N2-r:N2+r] = 0.25
    
    return A, B

def draw(A,B):
   
    fig, ax = pl.subplots(1,2,figsize=(5.65,4))
    
    ax[0].imshow(A, cmap='Greys')
    ax[1].imshow(B, cmap='Greys')
    
    ax[0].set_title('A')
    ax[1].set_title('B')
    
    ax[0].axis('off')
    ax[1].axis('off')
```

```python
DA, DB, f, k = 0.14, 0.06, 0.035, 0.065 # bacteria
A, B = initial(200)

for t in range(N_simulation_steps):
    A, B = gray_step(A, B, DA, DB, f, k, delta_t)
    
draw(A,B)
```
